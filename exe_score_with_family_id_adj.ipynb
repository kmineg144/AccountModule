## Process score data

```python
from score import Score
from params import subset_vars_acc, keeping_vars_sco

output_filepath = 'results_11_2020'

summary_csv =  os.path.join('data','summary_11_2020.csv')
acc_df = pd.read_pickle(os.path.join('outputs','account.pkl'))
```

```python
sco = Score(summary_csv)
sco.filter() 
sco.merge_data(acc_df[subset_vars_acc])
sco.add_columns()
sco.keep_vars(keeping_vars_sco)
sco.output_to_file(output_filepath, 'outputs')
```

```python
twin_cities_assigned = pd.read_excel(os.path.join(output_filepath,'twin_cities.xlsx'),
                            sheet_name='twin_cities_assigned')

twin_cities_excluded = pd.read_excel(os.path.join(output_filepath,'twin_cities.xlsx'),
                            sheet_name='twin_cities_excluded')

twin_cities_no_sfdc = pd.read_excel(os.path.join(output_filepath,'twin_cities.xlsx'),
                          sheet_name='twin_cities_no_sfdc')
```
